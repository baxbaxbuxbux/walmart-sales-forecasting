```python
import wandb
import pandas as pd
from neuralforecast import NeuralForecast

api = wandb.Api()

artifact_path = "grtsir22-free-university-of-tbilisi-/walmart-sales-forecasting/best-dlinear-model-final1:v0"

print("Downloading model from W&B...")
artifact = api.artifact(artifact_path)
model_dir = artifact.download(root="./champion_model_dir")

print(f"Artifact downloaded successfully to: {model_dir}")

nf = NeuralForecast.load(path=model_dir)
print("Model loaded into NeuralForecast successfully.")


print("Generating predictions...")
forecasts_df = nf.predict()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
train = pd.read_csv('train.csv')
```

```python
import pandas as pd
import numpy as np
from google.colab import files

test_df = test.copy()
test_df['Id'] = (
    test_df['Store'].astype(str) + '_' +
    test_df['Dept'].astype(str) + '_' +
    test_df['Date'].astype(str)
)

test_df['unique_id'] = test_df['Store'].astype(str) + '_' + test_df['Dept'].astype(str)

test_df['ds'] = pd.to_datetime(test_df['Date'])
forecasts_df['ds'] = pd.to_datetime(forecasts_df['ds'])

merged_df = pd.merge(test_df, forecasts_df, on=['unique_id', 'ds'], how='left')

merged_df['NBEATS'] = pd.to_numeric(merged_df['NBEATS'], errors='coerce')

historical_medians = train.groupby(['Store', 'Dept'])['Weekly_Sales'].median().reset_index()
historical_medians.rename(columns={'Weekly_Sales': 'Median_Sales'}, inplace=True)

merged_df = pd.merge(merged_df, historical_medians, on=['Store', 'Dept'], how='left')

merged_df['NBEATS'] = merged_df['NBEATS'].fillna(merged_df['Median_Sales']).fillna(0)

submission = pd.DataFrame({
    'Id': merged_df['Id'],
    'Weekly_Sales': merged_df['NBEATS']
})

submission_path = 'submission.csv'
submission.to_csv(submission_path, index=False)

print(f"Submission file successfully generated with {len(submission)} rows.")
files.download(submission_path)
```
